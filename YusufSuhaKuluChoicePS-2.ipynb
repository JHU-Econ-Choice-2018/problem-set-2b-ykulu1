### 1) Population Growth and Dynamic Inefficiency in the OLG Model

#### a) Derive an equation for the evolution of capital per worker

 


\begin{eqnarray*}
u() = log() \quad and \quad F(K,L) &=& K^\epsilon L^{1-\epsilon}, \\
w_t  &=&  f(k_t) - k_t f'(k_t), \\
r_t &=& f'(k_t), \\
where \quad f(k_t) &=&  k_t^{\epsilon}, \\
w_t &=& k_t^{\epsilon}(1 - \epsilon), \\
r_t &=& \epsilon k_t^{\epsilon - 1}, \\
\frac{1}{c_{1,t}} &=& \beta R \frac{1}{c_{2,t+1}}, \quad \textrm{Euler Equation} \\
c_{1,t} &=& \frac{w_{1,t}}{1+\beta}, \quad \textrm{EE + Budget} \\
a_{1,t} &=& w_{1,t} - c_{1,t} = k_t^{\epsilon}(1-\epsilon)\big(\frac{\beta}{1+\beta}\big), \quad \textrm{savings}\\
k_{t+1} &=& k_t^{\epsilon}\Big(\frac{1-\epsilon}{\Xi_{t+1}}\Big)\Big(\frac{\beta}{1+\beta}\Big)
\end{eqnarray*}

Note that savings of the today convert into tomorrow's capital. We divided the savings by the population growth factor because population tomorrow is $\Xi$ times of today and the "k" represents capital per worker. 

#### b) Derive the SS level of k$_t$

At the SS, k$_{t+1}$ = k$_t$ = $\bar{k}$
\begin{eqnarray}
\bar{k} = {\Bigg(\frac{1-\epsilon}{\Xi}\frac{\beta}{1+\beta}\Bigg)}^{\frac{1}{1-\epsilon}}
\end{eqnarray}

#### c) Will the new SS k associated with $\hat{\Xi}$ be larger or smaller than before?

\begin{eqnarray}
\bar{\bar{k}} = {\Bigg(\frac{1-\epsilon}{\hat{\Xi}}\frac{\beta}{1+\beta}\Bigg)}^{\frac{1}{1-\epsilon}} \quad where \quad \hat{\Xi} > \Xi
\end{eqnarray}

Note that it can be seen that the new steady state per worker capita level is lower. The intution is when there is a sudden increase in the population growth rate, each HH will end up with less capital level than before.

$\chi_t$ = $c_{1,t}$ + $\frac{c_{2,t}}{\Xi}$

This is an appropriate measure of consumption per person at period t. Suppose $X_t$ is the aggregate consumption at period t. Thus $X_t$ = $C_{1,t}$ + $C_{2,t}$. To find an aggregate consumption index per person in period t, we can either divide it to whole population which is $L_{t}$+$L_{t-1}$ or divide it to only young HHs population which is $L_{t}$ or we can divide it to old HH population which is $L_{t-1}$. Here $\chi_{t}$ is obtanied by dividing the aggregate consumption on the number of young HH, which is $L_{t}$. We obtain the equation. 

Notice that sustainable level of $\chi_{t}$ can be found as ;

\begin{eqnarray}
C_{1,t} + C_{2,t} + K_{t+1} - K_{t} &=& F(K,L) \quad \textrm{divide by $L_{t}$}, \\
\chi_{t}+\Xi k_{t+1} &=& f(k_t)+k_{t}, \\
\chi_{t} &=& \frac{1+\epsilon \beta}{1+\beta}k_{t}^{\epsilon} + k_{t}
\end{eqnarray}

### 2) Social Security and the Baby Boom

The Baby Boom generation is a generation which is bigger than the generation before and after itself.

First, let us assume that the policy keeps per capita benefit constant across generations and there is Pay As You Go social security system. This policy will have negative effects for the generation that comes after the "Baby Boom Generation" and the "Baby Boom Generation" will benefit such a system. They will be paying less taxes per person as their population is high, to keep the other generations's benefits constant, they need to give less taxes comparing with the other generations. When they get old, they will receive the same benefit per person comparing with the other generations but the taxes which are taken from the young generations will be much higher comparing with the other generations. 

Now let us assume that the policy keeps per capita taxes constant across generations. This will benefit the generation before the "Baby Boom Generation" and will have negative effects for the "Baby Boom Generation". When the Baby Boom generation is young, they are taxed same with the all other generations. Accumulating their taxes will be high as their population is high. Hence, when we subsidy the generation before the "Baby Boom", their benefir per person will be much higher comparing with the other generations. When the "Baby Boom Generation" gets old, they will receive fewer subsidy per person, as the accumulated taxes collected from the generation after them is fewer comparing to them. 

To sum up, the constant benefit policy worse off the generation after the "Baby Boom" and better off "Baby Boom Generation". The constant tax policy worse off the "Baby Boom Generation" and makes better off the generation that comes before them.
